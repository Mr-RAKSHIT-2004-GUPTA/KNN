```python
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000,n_features=3,n_classes=2,random_state=42,n_redundant=0)
```

```python
X.shape , y.shape
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(f"accuracy : {accuracy_score(y_test,y_pred)}")
print(f"confusion matrix : \n{confusion_matrix(y_test,y_pred)}")
print(f"classification report : \n{classification_report(y_test,y_pred)}")
```

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
param = {
    'n_neighbors' : [3,5,6,7,9,11,13],
    'weights' : ['uniform','distance'],
    'algorithm' : ['auto','ball_tree','kd_tree','brute'],
    'leaf_size' : [10,20,30,40],
    'metric' : ['euclidean','manhattan']
}
```

```python
grid_search = GridSearchCV(estimator=clf,param_grid=param,cv=5,n_jobs=-1,verbose = 3)
```

```python
grid_search
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
y_pred = grid_search.predict(X_test)
```

```python
print(f"accuracy : {accuracy_score(y_test,y_pred)}")
```

## KNN Regressor model

```python
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
X,y = make_regression(n_samples=1000,n_features=2,noise=3,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
clf = KNeighborsRegressor()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(f"r2_score : {r2_score(y_test,y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {
    'n_neighbors' : [3,5,6,7,9,11,13],
    'weights' : ['uniform','distance'],
    'algorithm' : ['auto','ball_tree','kd_tree','brute'],
    'leaf_size' : [10,20,30,40],
    'metric' : ['euclidean','manhattan']
}
```

```python
grid = GridSearchCV(estimator=clf,param_grid=param,cv=5,n_jobs=-1,verbose = 3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_ , grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(f"r2_score : {r2_score(y_test,y_pred)}")
```

```python
best_model = grid.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
print(f"r2_score : {r2_score(y_test,y_pred)}")
```
